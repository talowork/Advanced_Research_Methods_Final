```python
import pandas as pd
import glob
```

```python
meta = pd.read_csv("Meta.csv")
meta
```

```python
hr_files = sorted(glob.glob("HR*.csv"))

dfs = []

for file in hr_files:
    df = pd.read_csv(file)
    
    # subject_id üret
    subject_num = file.replace("HR", "").replace(".csv", "")
    subject_id = f"P{subject_num.zfill(2)}"
    
    df["UID"] = subject_id
    
    dfs.append(df)

hr_all = pd.concat(dfs, ignore_index=True)
hr_all.head(50)
```

```python
hr_all.columns = [c.lower() for c in hr_all.columns]

# Olası isimleri normalize et
if "hr" in hr_all.columns:
    hr_all = hr_all.rename(columns={"hr": "heart_rate"})
```

```python
df = hr_all.merge(meta, on="UID", how="left")
df.head()
```

```python
print("hr_all columns:", hr_all.columns.tolist())
print("meta columns:", meta.columns.tolist())
```

```python
# uid kolonunu UID yap
hr_all = hr_all.rename(columns={"uid": "UID"})

# merge
df = hr_all.merge(meta, on="UID", how="left")

df.head()
```

```python
df["UID"] = df["UID"].astype(str).str.upper().str.strip()
df["UID"] = df["UID"].str.replace(r"^P(\d)$", r"P0\1", regex=True)  # P2 -> P02

meta["UID"] = meta["UID"].astype(str).str.upper().str.strip()

df = df.drop(columns=["AGE","GENDER","TSST","SSST","GalaxyWatch"], errors="ignore").merge(meta, on="UID", how="left")
df.head()
```

```python
# sadece gerekli kolonlar (meta + hr)
keep = ["UID", "timestamp", "heart_rate", "AGE", "GENDER", "TSST", "SSST", "GalaxyWatch"]
df2 = df[keep].copy()

df2.head()
```

```python
df2["timestamp"] = pd.to_numeric(df2["timestamp"], errors="coerce")
df2["heart_rate"] = pd.to_numeric(df2["heart_rate"], errors="coerce")

df2.isna().mean()
```

```python
df2 = df2.dropna(subset=["timestamp", "heart_rate"])

# 0 değerleri büyük ihtimalle invalid/boş ölçüm
df2 = df2[df2["heart_rate"] > 0]

# fizyolojik aralık filtresi
df2 = df2[(df2["heart_rate"] >= 30) & (df2["heart_rate"] <= 220)]

df2.describe()
```

```python
# milisaniye mi saniye mi otomatik anlayalım
unit = "ms" if df2["timestamp"].median() > 1e11 else "s"
df2["datetime"] = pd.to_datetime(df2["timestamp"], unit=unit, errors="coerce")

df2 = df2.dropna(subset=["datetime"])
df2 = df2.sort_values(["UID", "datetime"])

df2[["UID","datetime","heart_rate"]].head()
```

```python
df2.to_csv("galaxywatch_hr_clean_final.csv", index=False)
```
